```python
data['Price in $'] = data['MSRP'].apply(lambda x : x.split()[1])
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
# Removing null values in "int" values
data['Cylinders'].fillna(data['Cylinders'].mean(),inplace= True)
data['Length'].fillna(data['Length'].mean(), inplace=True)
data['EngineSize'].fillna(data['EngineSize'].mean(), inplace=True)
data['Horsepower'].fillna(data['Horsepower'].mean(), inplace=True)
data['MPG_City'].fillna(data['MPG_City'].mean(), inplace=True)
data['Weight'].fillna(data['Weight'].mean(), inplace=True)
data['MPG_Highway'].fillna(data['MPG_Highway'].mean(), inplace=True)
data['Wheelbase'].fillna(data['Wheelbase'].mean(), inplace=True)
```

```python
# fillna with not words
data['MSRP'].fillna("Not Available",inplace=True)
data['Make'].fillna("Not Available", inplace=True)
data['Model'].fillna("Not Available", inplace=True)
data['Origin'].fillna("Not Available", inplace=True)
data['DriveTrain'].fillna("Not Available", inplace=True)
data['Invoice'].fillna("Not Available", inplace=True)
data['Type'].fillna("Not Available", inplace=True)
```
