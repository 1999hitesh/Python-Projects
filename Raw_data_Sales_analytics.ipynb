```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Raw Sales Data.csv")
```

```python
data
```

```python
data.info()
```

```python
data["Qty"] = data["Qty"].str.replace(',','').astype(float)
```

```python
data
```

```python
data['Sales'] = data["Sales"].str.replace(',','').astype(float)
```

```python
data
```

```python
data['Month'] = pd.to_datetime(data['Month'])
```

```python
data['Sales Month'] = data['Month'].dt.month_name(locale = 'English')
data
```

```python
data["Product"] = data["Product Name"].apply(lambda x: re.split('(\d+)',x)[0])
data
```

### Monthwise number of quantities sold

```python
data1=pd.DataFrame(data.groupby(data["Sales Month"])["Qty"].sum()).reset_index()
data1
```

### Category wise Qty sold

```python
data2=pd.DataFrame(data.groupby(data["Category"])["Qty"].sum()).reset_index()
data2
```

### Sub-channel wise total sales

```python
# data["Total Sales"] = data["Qty"]*data["Sales"]
# data
```

```python
data3 = pd.DataFrame(data.groupby(data["Sub-Channels"])["Sales"].sum()).reset_index()
data3
```

### Highest sale in which month


```python
data4=pd.DataFrame(data.groupby(data["Sales Month"])["Sales"].sum()).reset_index().max()
data4
```
